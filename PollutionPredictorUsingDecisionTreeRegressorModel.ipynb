```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('/content/sample_data/delhi_aqi.csv')
data
```

```python
data.head()
```

```python
data.tail()
```

```python
col_use=['no','no2','o3','so2','pm2_5','pm10','nh3','co']
data=data[col_use]
```

```python
X = data.drop("co", axis=1) # predictor variables
y = data["co"] # target variable
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
# Train the decision tree regressor model
reg = DecisionTreeRegressor().fit(X_train, y_train)


# Make predictions on the testing set
y_pred = reg.predict(X_test)



# Evaluate the performance of the model using metrics
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
print("training score is",reg.score(X_train,y_train))
print("testing score is",reg.score(X_test,y_test))
```

```python
new_data = [[2.18,70.6,13.59,38.62,364.61,411.73,28.63]]
```

```python
new_pred = reg.predict(new_data)
mgpm3=new_pred/1000
print("Predicted CO value:", new_pred," Âµg/mÂ³")
if mgpm3>=0.5 and mgpm3<=2:
    print("Air quality is good..")
elif mgpm3>2 and mgpm3<10:
    print("Air quality is moderate")
elif mgpm3>=10:
    print("Air quality is dangerous")
else:
    print("Air quality is excellent")
```
